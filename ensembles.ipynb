# Formato para el producto basado en netcdf4.

Es más sencillo de lo que pensábamos en un primer momento.

Tenemos tres opciones;

* Añadir una **dimensión** sin más.
* Organizar los datos en **grupos**.
* Organizar los datos en **distintos ficheros netCDF** y tener un fichero extra con metadatos.

La tercera opción no la contemplo porque es mejor tener un solo fichero que varios.

La primera de las opciones es la más simple siempre que los datos sean **homogénemos**, de manera que el valor de las dimensiones coincidan.

Pegas:

* Tendremos que hacer un tratamiento previo de los datos interpolándolos para referirlos todos al mismo tiempo, espacio...
* Volumen de datos: seguramente el punto anterior genere más datos de los que teníamos inicialmente
    
La opción de los grupos es buena siempre que trabajemos por medio de nuestra API en caso de que los datos sean **heterogéneos**; no estén dados para el mismo valor de tiempo, espacio, etc...

Pegas:

* Acceso más lento: hay que hacer el procesado de datos on the fly.
* Menos transparente.

```python
# las dimensiones se definen como variables tambien
# se definen indicando la dependencia, el nombre y el tipo de variable
ensembles  = rootgrp.createVariable('ensemble', 'u1', ('ensemble'))
times      = rootgrp.createVariable('time', 'f8', ('time'))
latitudes  = rootgrp.createVariable('latitude','f4',('lat'))
longitudes = rootgrp.createVariable('longitude','f4',('lon'))

# definimos las variables con su dependencia
temp = rootgrp.createVariable('temp','f4',('ensemble','time','lat','lon'))     

# imprimimos las variables que tenemos en el dataset
print rootgrp.variables
```

```python
# ahora asignamos los valores a las variables
ensembles_values = np.arange(0,2,1, dtype=int)
ensembles[:] = ensembles_values

times_values = np.arange(0,6,1, dtype=int)
times[:] = times_values

latitude_values = np.arange(60.0,100.0,2.0)
latitudes[:] = latitude_values

longitude_values = np.arange(120.0,160.0, 2.0)
longitudes[:] = longitude_values
```

```python
# Ahora si quiero asignar unos valores a la variable temperatura
# tengo que ajustar el rango y especificar las dimensiones
# ensemble 0, tiempo 0, lat:, lon:
temp[0,0, :, :] = np.random.rand(20,20)

# relleno los datos de forma aleatoria en un bucle
for i in range(0,len(rootgrp.dimensions['ensemble'])):
    for j in range(0,len(rootgrp.dimensions['time'])):
        temp[i,j, :, :] = np.random.rand(20,20)
```
